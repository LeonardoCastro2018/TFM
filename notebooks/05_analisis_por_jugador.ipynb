```python
# 05_analisis_por_jugador.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar mtricas por equipo (solo si necesits comparaciones por pas)
df_equipos = pd.read_csv("C:/Users/Usuario/OneDrive/Documentos/Cursos/Sport Data Campus/Master en Python Avanzado al deporte/Modulo 11/Proyecto Final/Data/eventos_copa_america/metricas_por_equipo.csv")

# Cargar todos los eventos para anlisis por jugador
df = pd.read_csv("C:/Users/Usuario/OneDrive/Documentos/Cursos/Sport Data Campus\Master en Python Avanzado al deporte/Modulo 11/Proyecto Final/Data/eventos_copa_america/eventos_copa_america_2024.csv", low_memory=False)

# Verificar columnas
print("Columnas disponibles:")
print(df.columns.tolist())

# -------------------------------
# Top 15 jugadores con ms pases
# -------------------------------
top_pases = (
    df[df['type'] == 'Pass']
    .groupby(['player', 'team'])
    .size()
    .reset_index(name='cantidad_pases')
    .sort_values(by='cantidad_pases', ascending=False)
    .head(15)
)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_pases, x='cantidad_pases', y='player', hue='team', dodge=False)
plt.title('Top 15 Jugadores con Ms Pases - Copa Amrica 2024')
plt.xlabel('Cantidad de Pases')
plt.ylabel('Jugador')
plt.legend(title='Equipo')
plt.tight_layout()
plt.show()

# -------------------------------
# Top 15 jugadores con ms tiros
# -------------------------------
top_tiros = (
    df[df['type'] == 'Shot']
    .groupby(['player', 'team'])
    .size()
    .reset_index(name='cantidad_tiros')
    .sort_values(by='cantidad_tiros', ascending=False)
    .head(15)
)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_tiros, x='cantidad_tiros', y='player', hue='team', dodge=False)
plt.title('Top 15 Jugadores con Ms Tiros - Copa Amrica 2024')
plt.xlabel('Cantidad de Tiros')
plt.ylabel('Jugador')
plt.legend(title='Equipo')
plt.tight_layout()
plt.show()
```
