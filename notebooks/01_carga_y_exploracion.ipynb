```python
# 01_carga_y_exploracion.ipynb

# ✅ Carga inicial de datos de la Copa América 2024 (StatsBomb)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ruta al archivo consolidado
ruta = "../data/eventos_copa_america/eventos_copa_america_2024.csv"
df = pd.read_csv(ruta)

# Mostrar las primeras filas
print("\u2705 Primeras filas del dataset:")
display(df.head())

# Tamaño del dataset
print(f"\nTotal de filas: {df.shape[0]} | Total de columnas: {df.shape[1]}")

# Columnas disponibles
print("\nColumnas disponibles en el dataset:")
print(df.columns.tolist())

# Tipos de eventos registrados
print("\nTipos de eventos disponibles:")
print(df['type'].value_counts())

# Equipos presentes
print("\nEquipos presentes:")
print(df['team'].dropna().unique())

# Jugadores más participativos (por cantidad de eventos)
print("\nTop 10 jugadores con más eventos:")
print(df['player'].value_counts().head(10))

# Distribución de eventos por partido
if 'match_id' in df.columns:
    eventos_por_partido = df['match_id'].value_counts()
    plt.figure(figsize=(10, 5))
    sns.histplot(eventos_por_partido, bins=15, kde=False)
    plt.title("Distribución de eventos por partido")
    plt.xlabel("match_id")
    plt.ylabel("Cantidad de eventos")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Exportar resumen de tipos de eventos a CSV
resumen_eventos = df['type'].value_counts().reset_index()
resumen_eventos.columns = ['evento', 'frecuencia']
resumen_eventos.to_csv("../data/eventos_copa_america/resumen_tipos_eventos.csv", index=False)
print("\n✅ Archivo resumen_tipos_eventos.csv exportado correctamente.")
```
